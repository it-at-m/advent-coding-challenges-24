## Aufgabenstellung

In dieser ersten Challenge wollen wir herausfinden, welcher Weihnachtsmarkt in München insgesamt am längsten geöffnet hat.

Um den Datensatz zu laden und zu analysieren, werden wir das Python-Paket `pandas` verwenden.

`pandas` ist ein mächtiges Werkzeug, um Daten zu laden, zu analysieren und zu visualisieren.

Es besitzt eine exzellente Dokumentation, die auch für Anfänger gut verständlich ist: https://pandas.pydata.org/docs/getting_started/

Wir laden also erst einmal den Datensatz:

```python
# Pandas importieren
import pandas as pd

# Datensatz laden
df = pd.read_csv('weihnachtsmarkte-muenchen-2024-v2.csv')

# Datensatz anzeigen
df
```

## Aufgabe lösen

Nachdem wir den Datensatz geladen haben, wollen wir herausfinden, welcher Weihnachtsmarkt in München insgesamt am längsten geöffnet hat.

Dazu müssen wir die Spalten `Startdatum` und `Enddatum` in ein für uns nutzbares Format umwandeln und die Differenz berechnen.

Diese Differenz fügen wir als neue Spalte `Dauer` hinzu.

Danach müssen wir den Datensatz noch nach `Dauer` sortieren und den Weihnachtsmarkt mit der längsten Dauer ausgeben.

Zuletzt wollen wir die Weihnachtsmärkte und ihre Dauer in einem Diagramm visualisieren.

Viel Erfolg!

## Hinweise

Folgende Links könnten hilfreich sein:

- Wie man nur bestimmte Spalten eines Datensatzes auswählt: https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html#how-do-i-select-specific-columns-from-a-dataframe
- Wie man neue Spalten aus bestehenden Spalten berechnet: https://pandas.pydata.org/docs/getting_started/intro_tutorials/05_add_columns.html#how-to-create-new-columns-derived-from-existing-columns
- Wie man mit Zeitdaten umgeht: https://pandas.pydata.org/docs/getting_started/intro_tutorials/09_timeseries.html
- Wie man Tabellen sortiert: https://pandas.pydata.org/docs/getting_started/intro_tutorials/07_reshape_table_layout.html#sort-table-rows
- Wie man Diagramme erstellt: https://pandas.pydata.org/docs/getting_started/intro_tutorials/04_plotting.html

```python
# Filtere nur die Spalten 'Name', 'Startdatum' und 'Enddatum' heraus
df = df.filter(['Marktname', 'Startdatum', 'Enddatum'])

# Konvertiere 'Startdatum' und 'Enddatum' in das Datetime-Format
df['Startdatum'] = pd.to_datetime(df['Startdatum'], format='%d.%m.%Y')
df['Enddatum'] = pd.to_datetime(df['Enddatum'], format='%d.%m.%Y')

# Berechne die Dauer, die jeder Markt geöffnet ist, und füge sie als neue Spalte 'Dauer' hinzu
df['Dauer'] = (df['Enddatum'] - df['Startdatum']).dt.days

# Zeige den aktualisierten Datensatz an
df
```

```python
# Sortiere das DataFrame nach der Spalte 'Dauer' in absteigender Reihenfolge
df_sorted = df.sort_values(by='Dauer', ascending=False)

# Zeige den sortierten Datensatz an
df_sorted
```

```python
# Finde den Markt mit der längsten Dauer (erster Eintrag im sortierten DataFrame)
longest_open_market = df_sorted.iloc[0]

# Gib den Markt mit der längsten Dauer aus
print(f"Der Weihnachtsmarkt mit der längsten Dauer ist: {longest_open_market['Marktname']} mit {longest_open_market['Dauer']} Tagen.")
```

```python
# Diagramm erstellen
df_sorted.plot(x='Marktname', y='Dauer', kind='barh', figsize=(20, 8), title='Dauer der Weihnachtsmärkte in München')
```

## Bonus-Challenge für Fortgeschrittene: Visualisierung der Öffnungszeiten 

Als Fortgeschrittene*r kannst du versuchen, die Öffnungszeiten als parallele Zeitstrahle darzustellen.

Hierfür musst du mit der Bibliothek `matplotlib` ein komplexeres Diagramm erstellen.

Die Dokumentation von ``matplotlib`` findest du hier: https://matplotlib.org/stable/contents.html

```python
from datetime import datetime

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Sortiere das DataFrame nach der Spalte 'Startdatum' in aufsteigender Reihenfolge
df_sorted_by_start = df_sorted.sort_values(by='Startdatum', ascending=False)

# Bonus: Erstelle ein Diagramm, um die Märkte und ihre Dauer zu visualisieren
fig, ax = plt.subplots(figsize=(20, 8))

# Zeichne ein Balkendiagramm mit den Marktnamen und deren Start- und Enddatum
for index, row in df_sorted_by_start.iterrows():
    ax.plot([row['Startdatum'], row['Enddatum']], [row['Marktname'], row['Marktname']], marker='o')

# Markiere den 24. Dezember
ax.axvline(pd.to_datetime('2024-12-24'), color='red', linestyle='--', linewidth=2, label='Heiligabend (24.12.2024)')

# Markiere das aktuelle Datum
current_date = datetime.now()
ax.axvline(current_date, color='blue', linestyle='--', linewidth=2, label=f'Aktuelles Datum ({current_date.strftime("%d.%m.%Y")})')

# Passe das Diagramm an
ax.set_xlabel('Datum', fontsize=14)  # Setze die Beschriftung der x-Achse
ax.set_ylabel('Weihnachtsmarkt', fontsize=14)  # Setze die Beschriftung der y-Achse
ax.set_title('Öffnungszeiten der Weihnachtsmärkte in München', fontsize=16, fontweight='bold')  # Setze den Titel des Diagramms
ax.grid(True, linestyle='--', alpha=0.6)  # Füge ein Raster hinzu
ax.set_facecolor('whitesmoke')  # Setze die Hintergrundfarbe des Diagramms

# Setze die Hauptlocator auf Wochen mit Sonntagen
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.SU))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d.%m.%Y'))

# Passe die Schriftgröße der Achsenbeschriftungen an
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Zeige die Legende an
ax.legend()

# Zeige das Diagramm an
plt.tight_layout()
plt.show()
```
